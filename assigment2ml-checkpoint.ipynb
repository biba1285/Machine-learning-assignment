```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
columns = ['fLength', 'fWidth', 'fSize', 'fConc', 'fConc1', 'fAsym',
           'fM3Long', 'fM3Trans', 'fAlpha', 'fDist', 'class']
data = pd.read_csv('magic04.data', names=columns)

sns.countplot(data=data, x='class')
plt.title("Original Class Distribution")
plt.show()
```

```python
gamma = data[data['class'] == 'g']
hadron = data[data['class'] == 'h']
gamma_balanced = gamma.sample(len(hadron), random_state=42)
balanced_data = pd.concat([gamma_balanced, hadron]).sample(frac=1, random_state=42).reset_index(drop=True)

sns.countplot(data=balanced_data, x='class')
plt.title("Balanced Class Distribution")
plt.show()
```

```python
le = LabelEncoder()
balanced_data['class'] = le.fit_transform(balanced_data['class'])
X = balanced_data.drop('class', axis=1)
y = balanced_data['class']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)
```

```python
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)
y_pred_dt = dt_model.predict(X_test)
y_proba_dt = dt_model.predict_proba(X_test)[:, 1]
sns.heatmap(confusion_matrix(y_test, y_pred_dt), annot=True, fmt='d')
plt.title("Decision Tree - Confusion Matrix")
plt.show()
print(classification_report(y_test, y_pred_dt))
```

```python
fpr_dt, tpr_dt, _ = roc_curve(y_test, y_proba_dt)
plt.plot(fpr_dt, tpr_dt, label=f'AUC = {auc(fpr_dt, tpr_dt):.2f}')
plt.plot([0,1], [0,1], linestyle='--')
plt.title("ROC Curve - Decision Tree")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(20, 10))
plot_tree(dt_model, feature_names=X.columns, class_names=["Hadron", "Gamma"], filled=True, max_depth=3)
plt.title("Decision Tree (first 3 levels)")
plt.show()
```

```python
nb_model = GaussianNB()
nb_model.fit(X_train, y_train)
y_pred_nb = nb_model.predict(X_test)
y_proba_nb = nb_model.predict_proba(X_test)[:, 1]
sns.heatmap(confusion_matrix(y_test, y_pred_nb), annot=True, fmt='d')
plt.title("Naive Bayes - Confusion Matrix")
plt.show()
print(classification_report(y_test, y_pred_nb))
```

```python
fpr_nb, tpr_nb, _ = roc_curve(y_test, y_proba_nb)
plt.plot(fpr_nb, tpr_nb, label=f'AUC = {auc(fpr_nb, tpr_nb):.2f}', color='green')
plt.plot([0,1], [0,1], linestyle='--')
plt.title("ROC Curve - Naive Bayes")
plt.legend()
plt.show()
```

```python
rf = RandomForestClassifier(random_state=42)
grid_rf = GridSearchCV(rf, {'n_estimators': [50, 100, 150]}, cv=5, scoring='accuracy', n_jobs=-1)
grid_rf.fit(X_train, y_train)
best_rf = grid_rf.best_estimator_
y_pred_rf = best_rf.predict(X_test)
y_proba_rf = best_rf.predict_proba(X_test)[:, 1]
print(f"Best n_estimators: {grid_rf.best_params_['n_estimators']}")
sns.heatmap(confusion_matrix(y_test, y_pred_rf), annot=True, fmt='d')
plt.title("Random Forest - Confusion Matrix")
plt.show()
print(classification_report(y_test, y_pred_rf))
```

```python
fpr_rf, tpr_rf, _ = roc_curve(y_test, y_proba_rf)
plt.plot(fpr_rf, tpr_rf, label=f'AUC = {auc(fpr_rf, tpr_rf):.2f}', color='red')
plt.plot([0,1], [0,1], linestyle='--')
plt.title("ROC Curve - Random Forest")
plt.legend()
plt.show()
pd.Series(best_rf.feature_importances_, index=X.columns).sort_values(ascending=False).plot(kind='bar')
plt.title("Feature Importance - Random Forest")
plt.show()
```

```python
ada = AdaBoostClassifier(random_state=42)
grid_ada = GridSearchCV(ada, {'n_estimators': [50, 100, 150]}, cv=5, scoring='accuracy', n_jobs=-1)
grid_ada.fit(X_train, y_train)
best_ada = grid_ada.best_estimator_
y_pred_ada = best_ada.predict(X_test)
y_proba_ada = best_ada.predict_proba(X_test)[:, 1]
print(f"Best n_estimators: {grid_ada.best_params_['n_estimators']}")
sns.heatmap(confusion_matrix(y_test, y_pred_ada), annot=True, fmt='d')
plt.title("AdaBoost - Confusion Matrix")
plt.show()
print(classification_report(y_test, y_pred_ada))
```

```python
fpr_ada, tpr_ada, _ = roc_curve(y_test, y_proba_ada)
plt.plot(fpr_ada, tpr_ada, label=f'AUC = {auc(fpr_ada, tpr_ada):.2f}', color='purple')
plt.plot([0,1], [0,1], linestyle='--')
plt.title("ROC Curve - AdaBoost")
plt.legend()
plt.show()
pd.Series(best_ada.feature_importances_, index=X.columns).sort_values(ascending=False).plot(kind='bar')
plt.title("Feature Importance - AdaBoost")
plt.show()
```

```python
models = ['Decision Tree', 'Naive Bayes', 'Random Forest', 'AdaBoost']
acc = [accuracy_score(y_test, y_pred_dt), accuracy_score(y_test, y_pred_nb),
       accuracy_score(y_test, y_pred_rf), accuracy_score(y_test, y_pred_ada)]
prec = [precision_score(y_test, y_pred_dt), precision_score(y_test, y_pred_nb),
        precision_score(y_test, y_pred_rf), precision_score(y_test, y_pred_ada)]
rec = [recall_score(y_test, y_pred_dt), recall_score(y_test, y_pred_nb),
       recall_score(y_test, y_pred_rf), recall_score(y_test, y_pred_ada)]
f1s = [f1_score(y_test, y_pred_dt), f1_score(y_test, y_pred_nb),
       f1_score(y_test, y_pred_rf), f1_score(y_test, y_pred_ada)]
aucs = [auc(fpr_dt, tpr_dt), auc(fpr_nb, tpr_nb),
        auc(fpr_rf, tpr_rf), auc(fpr_ada, tpr_ada)]
df = pd.DataFrame({
    "Model": models,
    "Accuracy": acc,
    "Precision": prec,
    "Recall": rec,
    "F1 Score": f1s,
    "AUC": aucs
})
print(df)
df.set_index("Model").plot(kind='bar', figsize=(12, 6))
plt.title("Model Performance Comparison")
plt.ylim(0, 1.1)
plt.grid(True, linestyle='--')
plt.show()
```
